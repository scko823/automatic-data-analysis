In this notebook I will clean the data more from the public data generated by the previous notebook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv("data/public_data.csv", parse_dates= [1,2], index_col=0)
```

```python
# let take a look at the dataframe
data.head()
```

The first thing I do is to set the index to the date of the trip because this will be more informative than 0, 1.... 
Also I want to have the day of week column and hour column for hour of trip start time. Having these column will also help to do groupby operation later on

```python
data['date'] = data['Start Time'].map(lambda x: pd.Timestamp.date(x))
```

```python
data = data.set_index('date', drop=False) # should keep the date column
```

```python
data['Hour'] = data['Start Time'].map(lambda x: x.hour)
```

```python
day_dict={0:'Mon',1:'Tue',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
data['Day of week'] = data['Start Time'].map(lambda x: day_dict[x.dayofweek])
```

```python
data.head() # let look at the last few columns of the data frame
```

OK since I work Sun to Wed, it is showing that I drive a lot more for my workday than my weekend.

Lets look at it by the hour and by the day of week, and seperate the commute and other trips
we will first create two pivot tables the record how many trip I started by the hour and day of weeek

```python
other_pt=pd.pivot_table(data[data['Tags']=='Other'], index=['Day of week','Hour'],aggfunc='count')
other_pt = other_pt.ix[:, 'Average MPG']
other_pt.unstack().ix[['Sun','Mon','Tue','Wed','Thurs','Fri','Sat'],:]

commute_pt=pd.pivot_table(data[data['Tags']=='commute'], index=['Day of week','Hour'],aggfunc='count')
commute_pt = commute_pt.ix[:, 'Average MPG']
commute_pt.unstack().ix[['Sun','Mon','Tue','Wed','Thurs','Fri','Sat'],:]
```

Then lets create the heat maps to compare them.

```python
fig,axes= plt.subplots(2,2, figsize=(19,19))

weekday_comm=sns.heatmap(commute_pt.unstack().ix[['Sun','Mon','Tue','Wed'],:], linewidth=0.5, annot=True ,fmt=".0f" ,cmap='summer',ax=axes[0,0])
weekday_comm.set_title("weekday, count of 'COMMUTE' trip started by hour")

weekend_comm=sns.heatmap(commute_pt.unstack().ix[['Thurs','Fri','Sat'],:], linewidth=0.5, annot=True ,fmt=".0f" ,cmap='summer',ax=axes[1,0])
weekend_comm.set_title("weekend, count of 'COMMUTE' trip started by hour")
weekday_other=sns.heatmap(other_pt.unstack().ix[['Sun','Mon','Tue','Wed'],:], linewidth=0.5, annot=True ,fmt=".0f" ,cmap='spring',ax=axes[0,1])
weekday_other.set_title("weekday, count of 'OTHER' trip started by hour")
weekend_other=sns.heatmap(other_pt.unstack().ix[['Thurs','Fri','Sat'],:], linewidth=0.5, annot=True ,fmt=".0f" ,cmap='spring',ax=axes[1,1])

weekend_other.set_title("weekend, count of 'OTHER' trip started by hour")
plt.show()
```

We can infer a lot of interesting things from them.
1. On my OTHER trips, I had started trip at all hour of the day, while the latest I have stayed at work is 11pm (only three times from the data set. Not too bad).
2. for my weekend, I had a total of 8 trips under 'commute'. Thats from a while I was working mon-fri (about for a week I believe)
